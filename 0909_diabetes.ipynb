```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
db = pd.read_csv("/content/AI-class/diabetes.csv")
```

```python
input = db[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]].to_numpy()
target = db["Outcome"].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(input, target, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
dt = DecisionTreeClassifier(random_state=42)

rf = RandomForestClassifier(n_estimators=200, random_state=42)

lr = LogisticRegression(max_iter=500)
```

```python
dt.fit(train_input, train_target)
```

```python
rf.fit(train_input, train_target)
```

```python
lr.fit(train_input, train_target)
```

```python
dt_acc = accuracy_score(test_target, dt.predict(test_input))

rf_acc = accuracy_score(test_target, dt.predict(test_input))

lr_acc = accuracy_score(test_target, dt.predict(test_input))
```

```python
print(dt_acc)
print(rf_acc)
print(lr_acc)
```

```python
db = pd.read_csv("/content/AI-class/diabetes.csv")
```

```python
db.columns
```

```python
cols=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
db = pd.read_csv("/content/AI-class/diabetes.csv", header=0, names=cols).dropna()
```

```python
X = db.drop(columns=["Outcome"])

y = db["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X, y, test_size=0.2, stratify=None, random_state=42

)
```

```python
dt = DecisionTreeClassifier(random_state=42)

rf = RandomForestClassifier(n_estimators=200, random_state=42)

lr = LogisticRegression(max_iter=500)
```

```python
dt.fit(X_train, y_train)

rf.fit(X_train, y_train)

lr.fit(X_train, y_train)
```

```python
dt_acc = accuracy_score(y_test, dt.predict(X_test))

rf_acc = accuracy_score(y_test, rf.predict(X_test))

lr_acc = accuracy_score(y_test, lr.predict(X_test))
```

```python
print("=== Test Accuracy ===")

print(f"Decision Tree : {dt_acc:.4f}")

print(f"Random Forest : {rf_acc:.4f}")

print(f"Logistic Reg. : {lr_acc:.4f}")
```
