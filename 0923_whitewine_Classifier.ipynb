```python
import pandas as pd

# 파일 경로를 지정합니다.
file_path = 'winequality-white.csv'

# pd.read_csv를 사용하되, delimiter=';' 옵션을 추가하여
# 세미콜론을 기준으로 데이터를 나누도록 설정합니다.
df = pd.read_csv(file_path, delimiter=';')

# 데이터가 잘 나뉘었는지 확인하기 위해 상위 5개 행을 출력합니다.
print(df.head())
```

```python
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score

from sklearn.neighbors import KNeighborsClassifier
```

```python
df.columns
```

```python
df['quality'].value_counts()
```

```python
cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']

#여기서도 deliniter=';'를 지정하여 csv파일 내의 데이터를 ;를 기준으로 나눈다
df = pd.read_csv("winequality-white.csv",header=None, names=cols, delimiter=';').dropna()
```

```python
#타겟 데이터에 결측값이 있나 확인하기
df["quality"].value_counts()
```

```python
#결측치 확인하기
df.isnull().sum()
```

```python
#결측치를 제거하는 과정
df = df.dropna()
df.isnull().sum()
```

```python
# 모든 레이블을 숫자로 변형 -> 머신러닝은 문자로 된 데이터는 읽지 못한다.

from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df.columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
#타겟 데이터로 사용할 quality 제외
input = df.drop(columns=["quality"])

#quality을 타겟 데이터로 선택
target = df["quality"]
```

```python
target
```

```python
#훈련세트와 테스트 세트 분리
train_input, test_input, train_target, test_target = train_test_split(input, target, test_size=0.2, random_state=42)
```

```python
train_input.shape, test_input.shape, train_target.shape, test_target.shape
```

```python
#회귀 모델 생성(스케일링 적용하기)
#스케일링을 하기 전에 훈련세트와 테스트 세트를 먼저 분리해 주어야함
#특정 모델을 위해서 하는게 아니라 머신러닝과 딥러닝 모델에서는 기본적으로 스케일링을 진행함
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
#의사결정 나무 모델 생성
dt = DecisionTreeClassifier(random_state=42)
#랜덤 포레스트 모델 생성
rf = RandomForestClassifier(n_estimators=200, random_state=42)
#로지스틱 회귀 모델 생성
lr = LogisticRegression(max_iter=500)
#KNN 분류 모델 생성
kn = KNeighborsClassifier()
```

```python
#모델 학습 시키기
dt.fit(train_input, train_target)
rf.fit(train_input, train_target)
lr.fit(train_input, train_target)
kn.fit(train_input, train_target)
```

```python
#로지스틱 회귀 모델의 기울기와 절편 확인하기
print(lr.coef_, lr.intercept_)
```

```python
print(lr.coef_.shape, lr.intercept_.shape)
```

```python
#모델 예측 결과 생성

dt_pred = dt.predict(test_input)
rf_pred = rf.predict(test_input)
lr_pred = lr.predict(test_input)
kn_pred = kn.predict(test_input)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

dt_acc = accuracy_score(test_target, dt_pred)
rf_acc = accuracy_score(test_target, rf_pred )
lr_acc = accuracy_score(test_target, lr_pred)
kn_acc = accuracy_score(test_target, kn_pred)


print("=== DT Test Accuracy ===")
print(f"Decision Tree : {dt_acc:.4f}")
print (confusion_matrix(test_target,dt_pred))

print("=== RF Test Accuracy ===")
print(f"Random Forest : {rf_acc:.4f}")
print (confusion_matrix(test_target,rf_pred))

print("=== LR Test Accuracy ===")
print(f"Logistic Reg. : {lr_acc:.4f}")
print (confusion_matrix(test_target,lr_pred))

print("=== KN Test Accuracy ===")
print(f"Logistic Reg. : {lr_acc:.4f}")
print (confusion_matrix(test_target,kn_pred))
```
